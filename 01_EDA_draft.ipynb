# Analyse Exploratoire - Heart Disease Prediction

## 1. Import des bibliothèques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print('✅ Bibliothèques importées')
```

## 2. Chargement des données

```python
# Télécharger le dataset
!wget -q https://raw.githubusercontent.com/fedesoriano/heart-failure-prediction/master/heart.csv

df = pd.read_csv('heart.csv')
print(f"Shape: {df.shape}")
display(df.head())
```

## 3. Analyse des données

```python
print("=== Informations ===")
print(df.info())

print("\n=== Statistiques ===")
display(df.describe())

print("\n=== Valeurs manquantes ===")
print(df.isnull().sum())
```

```python
print("=== Distribution de la cible ===")
print(df['HeartDisease'].value_counts())
print(f"Pourcentage de malades: {df['HeartDisease'].mean()*100:.2f}%")
```

```python
# Visualisations
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
numeric_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak', 'FastingBS']
for i, col in enumerate(numeric_cols):
    row, col_idx = i // 3, i % 3
    df[col].hist(ax=axes[row, col_idx], bins=30, edgecolor='black', alpha=0.7)
    axes[row, col_idx].set_title(f'Distribution de {col}')
plt.tight_layout()
plt.show()
```

```python
# Matrice de corrélation
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Matrice de Corrélation')
plt.show()
```
